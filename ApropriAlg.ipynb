# Implementation of Apropri Algorithm

Apropri Algorithm is an algorithm for frequent item set mining and association rule. It proceeds by identifying the frequent individual items in the database and extending them to larger and larger item sets as long as those item sets appear sufficiently often in the database.

```python
import numpy as np 
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Loading the dataset using pandas

df = pd.read_excel('E:\CCBT-Term3\DataMining\ML-practice\Datasets\OnlineRetail.xlsx')
```

```python
#printing first five rows of dataset
df.head()
```

```python
#observing the dataset
df.info()
```

```python
#Checking for null values
df.isnull().sum()
```

```python
# Stripping extra spaces in the description 
df['Description'] = df['Description'].str.strip() 
  
# Dropping the rows without any invoice number 
df.dropna(axis = 0, subset =['InvoiceNo'], inplace = True) 
df['InvoiceNo'] = df['InvoiceNo'].astype('str') 
```

```python
# Searching for unique country names
df.Country.unique()
```

```python
# Splitting the data according to the region of transaction and making description as column names in 'basket_Italy'
# Transactions done in UK 
basket_France = (df[df['Country'] =="France"] 
          .groupby(['InvoiceNo', 'Description'])['Quantity'] 
          .sum().unstack().reset_index().fillna(0) 
          .set_index('InvoiceNo'))
```

```python
basket_France.head()
```

```python
# Applying onehot encoding so that data can be encoded to use in association models
def hot_encode(x): 
    if(x<= 0): 
        return 0
    if(x>= 1): 
        return 1
    
basket_France_encoded = basket_France.applymap(hot_encode) 
basket_France.head()
```

We can observe that all the numeric data is converted into 1 and 0's remained as is indicating purchased or not

```python
# Building aprori model 
freq_items = apriori(basket_France_encoded, min_support = 0.1, use_colnames = True) 
  
# Collecting the inferred rules in a dataframe 
rules = association_rules(freq_items, metric ="lift", min_threshold = 1) 
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
```
